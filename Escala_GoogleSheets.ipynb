# 📅 Escala Automática a partir do Google Sheets

Este notebook lê automaticamente a planilha publicada no Google Sheets (formato CSV),
mostra a escala em tabela e permite exportar para HTML para compartilhar via link.

A planilha é atualizada todos os dias às 8h — basta abrir/rodar este notebook para ver os dados mais recentes.

```python
import pandas as pd

# 🔗 Link CSV publicado da sua planilha
csv_url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vTtu-b1fJg3AaYjKpcmYvdB1AaZpWGTRPr76mBxyBczreE69_A_3_NLJ4OPuGMtefWyTzl56G0oklFo/pub?output=csv"

# Lendo os dados direto do Google Sheets
df = pd.read_csv(csv_url)

df.head()
```

## 🔍 Visualização completa da escala

```python
df
```

## 💾 Exportar para HTML (para virar um link/app web)
